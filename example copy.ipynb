```python
from Library import Algorithm, Cloud, Manifolds
import numpy as np
import math

import matplotlib.pyplot as plt
```

```python
L1 = 17.3
L2 = 7.8
L3 = 4.5


def inverse(x, t):
    y = np.zeros((4, 6))
    y[0] = x.copy()
    y[1] = x.copy()
    y[2] = x.copy()
    y[3] = x.copy()


    if(not np.isnan(x[0])):

        o41 = np.sqrt(1 - x[0])
        o42 = -np.sqrt(1 - x[0])

        y[0, 3] = o41
        y[1, 3] = o42
        y[2, 3] = o41
        y[3, 3] = o42

        ang1 = np.arctan2(o41, x[0])
        ang2 = np.arctan2(o42, x[0])

        x_1 = t[0] - L1*math.cos(ang1)
        y_1 = t[1] - L1*math.sin(ang1)
        x_2 = t[0] - L1*math.cos(ang2)
        y_2 = t[1] - L1*math.sin(ang2)


        if(abs(x_1**2 + y_1**2 - L2**2 - L3 ** 2) / (2*L2*L3) <= 1):

            x_ = x_1
            y_ = y_1

            q21 = np.arccos((x_**2 + y_**2 - L2**2 - L3 ** 2) / (2*L2*L3))
            q11 = np.arctan2(y_, x_) - np.arctan2(L3*math.sin(q21), L2 + L3*math.cos(q21))

            q22 = -q21
            q12 = np.arctan2(y_, x_) + np.arctan2(L3*math.sin(q22), L2 + L3*math.cos(q22))

            y[0, 1] = q11
            y[0, 2] = q11 + q21

            y[1, 1] = q12
            y[1, 2] = q12 + q22

        if(abs(x_2**2 + y_2**2 - L2**2 - L3 ** 2) / (2*L2*L3) <= 1):

            x_ = x_2
            y_ = y_2

            q21 = np.arccos((x_**2 + y_**2 - L2**2 - L3 ** 2) / (2*L2*L3))
            q11 = np.arctan2(y_, x_) - np.arctan2(L3*math.sin(q21), L2 + L3*math.cos(q21))

            q22 = -q21
            q12 = np.arctan2(y_, x_) + np.arctan2(L3*math.sin(q22), L2 + L3*math.cos(q22))

            y[2, 1] = q11
            y[2, 2] = q11 + q21

            y[3, 1] = q12
            y[3, 2] = q12 + q22


        
    elif(not np.isnan(x[1])):
        o51 = np.sqrt(1 - x[1])
        o52 = -np.sqrt(1 - x[1])

        y[0, 4] = o51
        y[1, 4] = o51
        y[2, 4] = o52
        y[3, 4] = o52

        ang1 = np.arctan2(o51, x[1])
        ang2 = np.arctan2(o52, x[1])

        x_1 = t[0] - L1*math.cos(ang1)
        y_1 = t[1] - L1*math.sin(ang1)
        x_2 = t[0] - L1*math.cos(ang2)
        y_2 = t[1] - L1*math.sin(ang2)

        if(abs(x_1**2 + y_1**2 - L1**2 - L3 ** 2) / (2*L1*L3) <= 1):
            
            x_ = x_1
            y_ = y_1

            q21 = np.arccos((x_**2 + y_**2 - L1**2 - L3 ** 2) / (2*L1*L3))
            q11 = np.arctan2(y_, x_) - np.arctan2(L3*math.sin(q21), L1 + L3*math.cos(q21))

            q22 = -q21
            q12 = np.arctan2(y_, x_) + np.arctan2(L3*math.sin(q22), L1 + L3*math.cos(q22))

            y[0, 2] = q11
            y[0, 0] = q11 + q21

            y[1, 2] = q12
            y[1, 0] = q12 + q22


        if(abs(x_2**2 + y_2**2 - L1**2 - L3 ** 2) / (2*L1*L3) <= 1):

            x_ = x_2
            y_ = y_2

            q21 = np.arccos((x_**2 + y_**2 - L1**2 - L3 ** 2) / (2*L1*L3))
            q11 = np.arctan2(y_, x_) - np.arctan2(L3*math.sin(q21), L1 + L3*math.cos(q21))

            q22 = -q21
            q12 = np.arctan2(y_, x_) + np.arctan2(L3*math.sin(q22), L1 + L3*math.cos(q22))

            y[2, 2] = q11
            y[2, 0] = q11 + q21

            y[3, 2] = q12
            y[3, 0] = q12 + q22

    elif(not np.isnan(x[2])):

        o61 = np.sqrt(1 - x[2])
        o62 = -np.sqrt(1 - x[2])

        y[0, 5] = o61
        y[1, 5] = o61
        y[2, 5] = o62
        y[3, 5] = o62

        ang1 = np.arctan2(o61, x[2])
        ang2 = np.arctan2(o62, x[2])

        x_1 = t[0] - L1*math.cos(ang1)
        y_1 = t[1] - L1*math.sin(ang1)
        x_2 = t[0] - L1*math.cos(ang2)
        y_2 = t[1] - L1*math.sin(ang2)

        if(abs(x_1**2 + y_1**2 - L1**2 - L2 ** 2) / (2*L1*L2) <= 1):

            x_ = x_1
            y_ = y_1

            q21 = np.arccos((x_**2 + y_**2 - L1**2 - L2 ** 2) / (2*L1*L2))
            q11 = np.arctan2(y_, x_) - np.arctan2(L2*math.sin(q21), L1 + L2*math.cos(q21))

            q22 = -q21
            q12 = np.arctan2(y_, x_) + np.arctan2(L2*math.sin(q22), L1 + L2*math.cos(q22))

            y[0, 0] = q11
            y[0, 1] = q11 + q21

            y[1, 0] = q12
            y[1, 1] = q12 + q22

        if(abs(x_2**2 + y_2**2 - L1**2 - L2 ** 2) / (2*L1*L2) <= 1):

            x_ = x_2
            y_ = y_2

            q21 = np.arccos((x_**2 + y_**2 - L1**2 - L2 ** 2) / (2*L1*L2))
            q11 = np.arctan2(y_, x_) - np.arctan2(L2*math.sin(q21), L1 + L2*math.cos(q21))

            q22 = -q21
            q12 = np.arctan2(y_, x_) + np.arctan2(L2*math.sin(q22), L1 + L2*math.cos(q22))

            y[2, 0] = q11
            y[2, 1] = q11 + q21

            y[3, 0] = q12
            y[3, 1] = q12 + q22



    elif(not np.isnan(x[3])):
        
        o11 = np.sqrt(1 - x[3])
        o12 = -np.sqrt(1 - x[3])

        y[0, 0] = o11
        y[1, 0] = o12
        y[2, 0] = o11
        y[3, 0] = o12

        ang1 = np.arctan2(x[3], o11)
        ang2 = np.arctan2(x[3], o12)

        x_1 = t[0] - L2*math.cos(ang1)
        y_1 = t[1] - L2*math.sin(ang1)
        x_2 = t[0] - L2*math.cos(ang2)
        y_2 = t[1] - L2*math.sin(ang2)

        if(abs(x_1**2 + y_1**2 - L2**2 - L3 ** 2) / (2*L2*L3) <= 1):
                
            x_ = x_1
            y_ = y_1

            q31 = np.arccos((x_**2 + y_**2 - L2**2 - L3 ** 2) / (2*L2*L3))
            q21 = np.arctan2(y_, x_) - np.arctan2(L3*math.sin(q31), L2 + L3*math.cos(q31))

            q32 = -q31
            q22 = np.arctan2(y_, x_) + np.arctan2(L3*math.sin(q32), L2 + L3*math.cos(q32))

            y[0, 1] = q21
            y[0, 2] = q21 + q31

            y[1, 1] = q22
            y[1, 2] = q22 + q32

        if(abs(x_2**2 + y_2**2 - L2**2 - L3 ** 2) / (2*L2*L3) <= 1):

            x_ = x_2
            y_ = y_2

            q31 = np.arccos((x_**2 + y_**2 - L2**2 - L3 ** 2) / (2*L2*L3))
            q21 = np.arctan2(y_, x_) - np.arctan2(L3*math.sin(q31), L2 + L3*math.cos(q31))

            q32 = -q31
            q22 = np.arctan2(y_, x_) + np.arctan2(L3*math.sin(q32), L2 + L3*math.cos(q32))

            y[2, 1] = q21
            y[2, 2] = q21 + q31

            y[3, 1] = q22
            y[3, 2] = q22 + q32

    elif(not np.isnan(x[4])):

        o21 = np.sqrt(1 - x[4])
        o22 = -np.sqrt(1 - x[4])

        y[0, 1] = o21
        y[1, 1] = o21

    return y


def Jx(x, t):
    J = np.zeros((2, 3))

    J[0, 0] = - L1*math.sin(x[0])
    J[1, 0] = L1*math.cos(x[0])

    J[0, 1] = - L2*math.sin(x[1])
    J[1, 1] = L2*math.cos(x[1])

    J[0, 2] = - L3*math.sin(x[2])
    J[1, 2] = L3*math.cos(x[2])

    return J

def Jt(x, t):
    J = np.zeros((2, 2))
    J[0, 0] = 1
    J[1, 1] = 1

    return J
```
