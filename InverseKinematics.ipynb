```python
from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Getting multiple solutions from a single solution using taylor expansion
def taylor1st(x0, dx = 0.001):

    y = np.zeros(2)

    y[0] = dx

    # Example robot
    y[1] =  - (2*x0[0] + x0[1]) * dx / (x0[0]+3*x0[1]**2)

    # Circle
    #y[1] = - (x0[0]) * dx / (x0[1])

    #normalize y
    y = y / np.linalg.norm(y)
    y = y * dx

    return y
```

```python
# Example robot
initial_guess = np.array([-4, -3.09])

# Circle
#initial_guess = np.array([0, 1])
```

```python
sol = initial_guess

points = np.zeros((10000,2))

for i in range(10000):
    sol = sol + taylor1st(sol, 0.003)
    
    points[i] = sol

plt.axis('equal')
plt.axis([-5, 5, -5, 5])





plt.plot(points[:,0], points[:,1], 'b-')

plt.show()

```
